```python
import pandas as pd
import numpy as np
from som import SOM
df = pd.read_csv('../data/iris.csv')
s = {'Iris-setosa': 0,'Iris-versicolor': 1,'Iris-virginica': 2}
df['class'] = df['class'].map(s)
targets = df['class'].to_numpy()
data = df[['sepal length','sepal width','petal length','petal width']].to_numpy()
data.shape
```

```python
som = SOM(10, 10)  # initialize the SOM
som.fit(data, 10000, save_e=True, interval=100)  # fit the SOM for 10000 epochs, save the error every 100 steps
som.plot_error_history(filename='som_error.png')  # plot the training error history
```

```python
# now visualize the learned representation with the class labels
som.plot_point_map(data, targets, ['Iris-setosa','Iris-versicolor','Iris-virginica'], filename='som.png')
som.plot_class_density(data, targets, t=0, name='setosa', filename='setosa.png')
som.plot_class_density(data, targets, t=1, name='versicolor', filename='versicolor.png')
som.plot_class_density(data, targets, t=2, name='virginica', filename='virginica.png')
som.plot_distance_map(filename='distance_map.png')  # plot the distance map after training
```
